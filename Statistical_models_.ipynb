```python
import pandas as pd
import statsmodels.api as sm

# Sample dataset structure, modify according to your actual dataset
df = df_merged.copy()

# Combine the ideology columns into a single binary indicator for each ideology
ideologies = ['environmentalism', 'feminism', 'islamism', 'kemalism', 'kurdish_national_movement','conservatism', 
              'liberalism', 'secularism', 'social_democracy', 'socialism', 'turkish_nationalism']

# Create a new binary column for each ideology that indicates if it is present in any of the three ideology columns
for ideology in ideologies:
    df[ideology] = ((df['ideology_1'] == ideology) | 
                    (df['ideology_2'] == ideology) | 
                    (df['ideology_3'] == ideology)).astype(int)

# Now, X will include all ideology columns
X = df[['unemployment_rate', 'gender', 'age_group'] + ideologies]

# Convert categorical variables to dummy variables (one-hot encoding), drop first level for reference category
X = pd.get_dummies(X, drop_first=True)

# Add a constant term for the intercept
X = sm.add_constant(X)
X=X.astype(float)

# Define the dependent variable
y = df['stance_immigration']

# Ensure the dependent variable is treated as categorical
y = y.astype('category')


```

```python

# Fit the multinomial logistic regression model
model = sm.MNLogit(y, X)
result = model.fit()

# Display the summary of the model
print(result.summary())
```

### Emotion Model

```python
# Define the mapping from Turkish to English emotion names
emotion_translation = {
    'minnet': 'Gratitude',
    'ofke': 'Anger',
    'uzuntu': 'Sadness',
    'arzu': 'Desire',
    'onaylamama': 'Disapproval',
    'pismanlik': 'Regret',
    'igrenme': 'Disgust',
    'notr': 'Neutral',
    'saskinlik': 'Surprise',
    'umutsuzluk': 'Hopelessness',
    'umut': 'Hope',
    'sevgi': 'Love',
    'korku': 'Fear',
    'kaygi': 'Anxiety',
    'mutluluk': 'Happiness',
    'onaylama': 'Approval',
    'utanc': 'Shame'
}

# Function to process and replace emotions in the dataset
def translate_emotions(emotion_list):
    translated_emotions = [emotion_translation.get(emotion, emotion) for emotion in emotion_list]
    return translated_emotions

# Apply the function to the 'emotions' column
df_merged['emotions'] = df_merged['emotions'].apply(translate_emotions)

# Display the updated dataset
df_merged['emotions']
```

```python
import pandas as pd
import statsmodels.api as sm

# Sample dataset structure, modify according to your actual dataset
df = df_merged.copy()

# List of main emotions to consider
main_emotions = ['Happiness', 'Sadness', 'Anger', 'Fear', 'Anxiety', 'Disgust']

# Create a binary column for each main emotion
for emotion in main_emotions:
    df[emotion] = df['emotions'].apply(lambda x: 1 if emotion in x else 0)

# Now, X will include all main emotion columns
X = df[['unemployment_rate', 'gender', 'age_group'] + main_emotions]

# Convert categorical variables to dummy variables (one-hot encoding), drop first level for reference category
X = pd.get_dummies(X, drop_first=True)

# Add a constant term for the intercept
X = sm.add_constant(X)
X=X.astype(float)
# Define the dependent variable
y = df['stance_immigration']

# Ensure the dependent variable is treated as categorical
y = y.astype('category')

```

```python
# Fit the multinomial logistic regression model
model = sm.MNLogit(y, X)
result = model.fit()

# Display the summary of the model
print(result.summary())
```

##  Interaction Model

```python
import statsmodels.api as sm
import pandas as pd

df1 = df_merged.copy()

# Assuming 'df1' is your dataset and it includes the emotion and ideology presence columns
main_emotions = ['Happiness', 'Sadness', 'Anger', 'Fear', 'Anxiety', 'Disgust']

for emotion in main_emotions:
    df1[emotion] = df1['emotions'].apply(lambda x: 1 if emotion in x else 0)

# Create binary columns for ideology presence (1 if the ideology is present, 0 otherwise)
ideologies = ['islamism', 'kemalism', 'conservatism','feminism',"environmentalism",'kurdish_national_movement',
              'liberalism', 'secularism', 'social_democracy', 'socialism', 'turkish_nationalism']

for ideology in ideologies:
    df1[f'ideology_presence_{ideology}'] = (
        (df1['ideology_1'] == ideology) |
        (df1['ideology_2'] == ideology) |
        (df1['ideology_3'] == ideology)
    ).astype(int)

# Create dummy variables for gender and age_group
df1 = pd.get_dummies(df1, columns=['gender', 'age_group'], drop_first=True)

# Create interaction terms between each emotion and ideology presence
interaction_terms = []
emotions = ['Happiness', 'Sadness', 'Anger', 'Fear', 'Anxiety', 'Disgust']


for emotion in emotions:
    for ideology in ideologies:
        interaction_term = f'{emotion}*ideology_presence_{ideology}'
        df1[interaction_term] = df1[emotion] * df1[f'ideology_presence_{ideology}']
        interaction_terms.append(interaction_term)


# Prepare the independent variables (X) and dependent variable (y)
X = df1[['unemployment_rate'] + list(df1.filter(like='gender_').columns) + list(df1.filter(like='age_group_').columns) +
       emotions + [f'ideology_presence_{ideology}' for ideology in ideologies] + interaction_terms]
X = sm.add_constant(X)  # Add a constant term for the intercept
y = df1['stance_immigration']
```

```python
model = sm.MNLogit(y, X)
result = model.fit(method='bfgs',maxiter=1000)

# Display the summary of the model
print(result.summary())
```

```python
import pandas as pd

# Assuming df_merged is your dataframe with 'big5' and 'welfare' columns containing lists

# Step 1: Create a set of unique topics in 'big5' and 'welfare'
unique_big5 = set([topic for sublist in df1['big5'] for topic in sublist])
unique_welfare = set([topic for sublist in df1['welfare'] for topic in sublist])

# Step 2: Create binary columns for each unique topic in 'big5'
for topic in unique_big5:
    df[f'big5_{topic}'] = df['big5'].apply(lambda x: 1 if topic in x else 0)

# Step 3: Create binary columns for each unique topic in 'welfare'
for topic in unique_welfare:
    df[f'welfare_{topic}'] = df['welfare'].apply(lambda x: 1 if topic in x else 0)


```

```python
import pandas as pd
import statsmodels.api as sm

# Assuming 'df' is your DataFrame
df1 = df.copy()

# List of emotions (binary columns already present in the dataset)
emotions = ['Sadness', 'Anger', 'Fear', 'Anxiety', 'Disgust']

# List of ideologies
ideologies = ['islamism', 'kemalism', 'conservatism', 'feminism', 'environmentalism',
              'kurdish_national_movement', 'liberalism', 'secularism',
              'social_democracy', 'socialism', 'turkish_nationalism']

# List of relevant topics
topics = ['big5_economy', 'big5_internal_affairs', 'welfare_health_and_public_health', 
          'welfare_social_policy', 'welfare_education']

# Convert ideologies to binary presence columns based on ideology_1, ideology_2, and ideology_3
for ideology in ideologies:
    df1[ideology] = ((df1['ideology_1'] == ideology) |
                     (df1['ideology_2'] == ideology) |
                     (df1['ideology_3'] == ideology)).astype(int)

# Create interaction terms and store them in a dictionary, then concatenate
interaction_data = {}

# Create two-way interactions between emotions and ideologies
for emotion in emotions:
    for ideology in ideologies:
        interaction_term = f'{emotion}*{ideology}'
        interaction_data[interaction_term] = df1[emotion] * df1[ideology]

# Create two-way interactions between emotions and topics
for emotion in emotions:
    for topic in topics:
        interaction_term = f'{emotion}*{topic}'
        interaction_data[interaction_term] = df1[emotion] * df1[topic]

# Create three-way interactions between emotions, ideologies, and topics
for emotion in emotions:
    for ideology in ideologies:
        for topic in topics:
            interaction_term = f'{emotion}*{ideology}*{topic}'
            interaction_data[interaction_term] = df1[emotion] * df1[ideology] * df1[topic]

# Create a DataFrame from the dictionary of interaction terms
interaction_df = pd.DataFrame(interaction_data)

# Concatenate the interaction terms to the original DataFrame
df1 = pd.concat([df1, interaction_df], axis=1)

# Prepare the independent variables (X) and dependent variable (y)
# Add all emotions, ideologies, topics, and interaction terms as features
X = df1[['unemployment_rate'] + emotions + ideologies + topics + list(interaction_df.columns)]

# Add gender and age group as dummy variables
X = pd.get_dummies(df1[['gender', 'age_group']], drop_first=True).join(X)

# Add a constant for the intercept
X = sm.add_constant(X)
X=X.astype(float)
# Define the dependent variable (stance on immigration)
y = df1['stance_immigration'].astype('category')

```
