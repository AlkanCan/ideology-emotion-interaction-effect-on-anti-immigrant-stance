```python
import pandas as pd

# Assuming df is your main DataFrame and it has been properly loaded

# Extracting 15 irrelevant tweets
df_neutral = df[df['stance'] == 0].sample(n=32, random_state=42)  # Adjust random_state for reproducibility

# Extracting 15 relevant tweets
df_positive = df[df['stance'] == 1].sample(n=32, random_state=42)  # Adjust random_state for reproducibility

df_negative=df[df['stance']==2].sample(n=32, random_state=42)
# Combining both samples into one DataFrame for easier handling
df_sample = pd.concat([df_neutral, df_positive,df_negative])

# Optionally, you can shuffle the rows if you want to mix the relevant and irrelevant tweets
df_sample = df_sample.sample(frac=1, random_state=42).reset_index(drop=True)

# Print or use df_sample as needed
df_sample
```

```python
import openai

openai.api_key = ''  # Securely stored API key


def classify_tweet(tweet):
    try:
        response = openai.ChatCompletion.create(
            model="gpt-4o",
            messages=[
                {"role": "system", "content": "You are an assistant trained to classify tweets about immigration in Turkey into three categories: positive, negative, or neutral. Provide a brief response."},
                {"role": "user", "content": f"Classify the following tweet related to immigration in Turkey into one of three categories: positive, negative, or neutral. The tweets may include negative sentiment but expressing positive stance toward immigrants. Label those as positive.Similarly can include negative or positive sentiment but does not express a clear stance toward immigration or immigrants, label those as neutral."
                                            "A tweet should be classified as:"
                                             "- Positive if it shows empathy or sympathy towards immigrant or refugee groups in Turkey, opposes racism and discrimination within the context of immigration, or discusses the advantages of immigrants and refugees or support them."
                                             "Example: 'suriyeden ülkemize gelmek zorunda kalmış insanlara karşı ırkçı ayrımcı yabancı düşmanlarının bizi yönlendirmesine izin vermeyelim.'"
                                             "Example: 'Hocam milyon Afgan gelirse işçilik ucuzlar ülkeye bereket gelir.'\n\n"
                                             "- Negative if it expresses opposition to immigration, criticizes immigrant or refugee groups, or criticizes the government's policies that lead to immigration, suggesting that these policies are problematic or harmful."
                                             "Example: 'kusura bakmada bundan yıl önce çocukken bende aynısını diyordum ben mülteci ülkeye gelen çocuklar büyüdü şimdi ülkede huzur bırakmadılar artık kadınına çocuğuna üzülmüyorum.'\n"
                                             "Example: 'O kadar göçmen geldi ki doğal artık sığınmacılar seviyemizi düşürdü malum şahıs yüzünden.'"
                                             "- Neutral if the tweet does not clearly express a negative or positive stance towards immigration, is merely informative, or unrelated to immigration issues in Turkey."
                                             "Example: 'ysk suriyeliler oy kullanamaz bakan soylu bin suriyeli oy kullanıcak kimin dediği belli değil.'"
                                             "Example: 'Adamın sadece Suriyeli problemi yok tamamen Türk olmayan herşeyle problemi var.'"
                                             "Just a brief label, nothing more"
                                             f"Tweet: {tweet}"} ],
            max_tokens=15  # Limiting the response length
        )
        return response['choices'][0]['message']['content']
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

# Example usage
```

```python
df_sample['ChatGPT'] = df_sample['Tweet'].apply(classify_tweet)

```

```python
df_sample['ChatGPT']=df_sample['ChatGPT'].replace({"Negative":2,'Positive':1,'Neutral':0,'Neutral.':0,"['positive']":1,'':0})
```
