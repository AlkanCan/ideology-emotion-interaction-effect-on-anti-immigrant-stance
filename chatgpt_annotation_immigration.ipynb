```python
df = pd.read_csv("preprocessed_stop_word_removed_saglama.csv")
```

```python
import pandas as pd

# Assuming df is your main DataFrame and it has been properly loaded

# Extracting irrelevant tweets
df_irrelevant = df[df['label'] == 0].sample(n=273, random_state=42)  # Adjust random_state for reproducibility

# Extracting relevant tweets
df_relevant = df[df['label'] == 1].sample(n=91, random_state=42)  # Adjust random_state for reproducibility

# Combining both samples into one DataFrame for easier handling
df_sample = pd.concat([df_irrelevant, df_relevant])

# Optionally, you can shuffle the rows if you want to mix the relevant and irrelevant tweets
df_sample = df_sample.sample(frac=1, random_state=42).reset_index(drop=True)

# Print or use df_sample as needed
df_sample
```

```python
df_sample=df_sample[['Tweet','label']]
```

```python
import openai

openai.api_key =   # Securely stored API key

# Load your dataset

def classify_tweet(tweet):
    try:
        response = openai.Completion.create(
            engine="gpt-3.5-turbo-instruct",
            prompt=(  "Classify the following tweet as 'relevant' or 'irrelevant' to discussions on immigration in Turkey. "
        "Consider a tweet relevant if it directly mentions immigrant groups such as Syrians, Afghans, or Ukrainians, discusses the impacts of immigration policies, debates societal attitudes towards immigrants, or refers to Turkey's international relations affecting immigration to Turkey. "
        "Here are some example tweets:"
        "Relevant: 'bursada kafa keseceÄŸim diyen onun bunun evladÄ± suriyeli ÅŸerefsiz gÃ¶zaltÄ±na alÄ±nmÄ±ÅŸ onu gÃ¶nderin kelle koltukta yaÅŸasÄ±n birazda'. "
        "Irrelevant: 'Ã¼lkeri kuran yahudilerdir Ã¼ller ailesi gÃ¶rÃ¼ntÃ¼dÃ¼r'. "
        "Pay special attention to keywords such as 'mÃ¼lteci', 'gÃ¶Ã§men', 'suriyeli', 'afgan','sÄ±ÄŸÄ±nmacÄ±' which are strong indicators of relevance to immigration. "
        "Based on the description and examples provided, classify the following tweet as either 'relevant' or 'irrelevant' to the topic of immigration in Turkey: '{}'"
        .format(tweet)),
            max_tokens=10
        )
        return response.choices[0].text.strip()
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

# Classify each tweet

```

```python
df_sample['ChatGPT'] = df_sample['Tweet'].apply(classify_tweet)

```
